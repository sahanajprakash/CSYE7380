#### INCOME STATEMENT:
1: Gross Margin
- Equation: Gross Profit / Revenue
- Rule: 40% or higher
- Buffett's Logic: Signals the company isn’t competing on price.

2: Selling General And Administration (SG & A Expense) Margin
- Equation: SG & A Expense / Gross Profit
- e.g. Selling General and Administration / Gross Profit
- Rule: 30% or lower
- Buffett's Logic: Wide-moat companies don’t need to spend a lot on overhead to operate.

3: Research And Development	(R & D Expense) Margin
- Equation: R & D Expense / Gross Profit
- e.g. Research And Development / Gross Profit
- Rule: 30% or lower
- Buffett's Logic: R&D expenses don't always create value for shareholders. 

4: Depreciation Margin
- Equation: Depreciation / Gross Profit
- e.g. Reconciled Depreciation / Gross Profit
- Rule: 10% or lower
- Buffett's Logic: Buffett doesn't like businesses that need to invest in depreciating assets to maintain their competitive advantage.

5: Interest Expense Margin
- Equation: Interest Expense / Operating Income
- Rule: 15% or lower
- Buffett's Logic: Great businesses don’t need debt to finance themselves.

6: Income Tax Expenses
- Equation: Tax Paid / Pre-Tax Income
- Tax Provision / Pre-Tax Income
- Rule: Current Corporate Tax Rate
- Buffett's Logic: Great businesses are so profitable that they are forced to pay their full tax load.

7: Net Margin (Profit Margin)
- Equation: Net Income / Sales
- e.g. Net Income / Total Revenue
- Rule: 20% or higher
- Buffett's Logic: Great companies convert 20% or more of their revenue into net income.

8: Earnings Per Share Growth
- Equation: Year 2 EPS / Year 1 EPS
- e.g. 09/30/2024 Basic EPS / 09/30/2023 Basic EPS
- Rule: Positive & Growing
- Buffett's Logic: Great companies increase profits every year.

```python
import pandas as pd
import numpy as np
import yfinance as yf

# Specify the stock you want to analyze (e.g., Medco)
stock_symbol = 'TSLA'
stocks = yf.Ticker('TSLA')
print(f'{stock_symbol}: FINANCIAL STATEMENT')
stocks.financials
```

```python
# Compute the ratios
financial_statement_readings = pd.DataFrame({
    'Gross Margin': stocks.financials.loc['Gross Profit'] / stocks.financials.loc['Total Revenue'], # 40% or higher
    'SG & A Expense Margin': stocks.financials.loc['Selling General And Administration'] / stocks.financials.loc['Gross Profit'], # 30% or lower
    'R & D Expense Margin': stocks.financials.loc['Research And Development'] / stocks.financials.loc['Gross Profit'], # 30% or lower
    'Depreciation Margin': stocks.financials.loc['Reconciled Depreciation'] / stocks.financials.loc['Gross Profit'], # 10% or lower
    'Interest Expense Margin': stocks.financials.loc['Interest Expense'] / stocks.financials.loc['Operating Income'],# 15% or lower
    'Income Tax Expense': stocks.financials.loc['Tax Provision'] / stocks.financials.loc['Pretax Income'], # Current Corporate Tax Rate
    'Profit Margin': stocks.financials.loc['Net Income'] / stocks.financials.loc['Total Revenue'], # 20% or higher
    'Earnings Per Share Growth': [stocks.financials.loc['Basic EPS', '2024-09-30'] / stocks.financials.loc['Basic EPS', '2023-09-30'], 
                                  stocks.financials.loc['Basic EPS', '2023-09-30'] / stocks.financials.loc['Basic EPS', '2022-09-30'], 
                                  stocks.financials.loc['Basic EPS', '2022-09-30'] / stocks.financials.loc['Basic EPS', '2021-09-30'],
                                  np.nan, np.nan] # > 1 Positive & Growing
}).T
print(financial_statement_readings)
```

#### BALANCE SHEET:
9: Cash & Debt
- Equation:  Cash > Debt
- e.g. Cash And Cash Equivalents / Current Debt
- Rule: More cash than debt
- Buffett's Logic: Great companies generate lots of cash without needing much debt. 

11: Adjusted Debt to Equity
- Equation:  Total Liabilities / Shareholder Equity + Treasury Stock
- Total Shareholder Equity is the residual value of assets after deducting total liabilities. It includes common stock, retained earnings, and additional paid-in capital. 
- e.g. Total Debt / (Total Asset - Total Debt)
- Rule : < 0.80
- Buffett's Logic: Great companies finance themselves with equity. 

12: Preferred Stock
- Rule: None
- Buffett's Logic: Great companies don't need to fund themselves with preferred stock. 

13: Retained Earnings
- Equation:  Year 1 / Year 2
- Rule:  Consistent growth
- Buffett's Logic: Great companies grow retained earnings each year. 

14: Treasury Stock
- Rule: Exists
- Buffett's Logic: Great companies repurchase their stock. 

```python
print(f'{stock_symbol}: BALANCE SHEET')
pd.set_option('display.max_rows', None)
stocks.balancesheet
```

```python
# Compute the ratios
balance_sheet_readings = pd.DataFrame({
    'Cash > Debt': stocks.balancesheet.loc['Cash And Cash Equivalents'] / stocks.balancesheet.loc['Current Debt'], # Rule: cash > debt
    'Adjusted Debt to Equity': stocks.balancesheet.loc['Total Debt'] / (stocks.balancesheet.loc['Total Assets'] - stocks.balancesheet.loc['Total Debt']) # Rule : < 0.80
}).T
print(balance_sheet_readings)
```

#### CASH FLOW STATEMENT:
15: Capital Expenditures (CapEx) Margin
- Equation:  CapEx / Net Income
- e.g., Capital expenditures / Net Income From Continuing Operations
- Rule: <25%
- Buffett's Logic: Great companies don't need much equipment to generate profits.

```python
print(f'{stock_symbol}: CASHFLOW STATEMENT')
stocks.cashflow
```

```python
# Compute the ratios
cashflow_statement_readings = pd.DataFrame({
    'CapEx Margin': -stocks.cashflow.loc['Capital Expenditure'] / stocks.cashflow.loc['Net Income From Continuing Operations'], # Rule: <25%
}).T
print(cashflow_statement_readings)
```
